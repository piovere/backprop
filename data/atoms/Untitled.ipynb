```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(wavelength, mean, std):
    fp = 1 / np.sqrt(2.0 * np.pi * std ** 2)
    ep = - (mean - wavelength) ** 2 / (2.0 * std ** 2)
    sp = np.exp(ep)
    return fp * sp
```

```python
def peak(wavelength, intensity, std, n_samples=1000000, min_wavelength=200, max_wavelength=1100):
    wavelengths = np.linspace(min_wavelength, max_wavelength, n_samples)
    weights = gaussian(wavelengths, wavelength, std)
    return intensity * weights
```

```python
wl = np.linspace(200, 1100, 1000000)
plt.plot(wl, peak(1000, 50, 10)+peak(300, 30, 8)+peak(580, 20, 10)+peak(600, 30, 10))
```
